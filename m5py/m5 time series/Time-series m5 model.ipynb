```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error
from m5py import M5Prime, export_text_m5
```

```python
df = pd.read_excel("D:\BTP\Gopalkheda and yerli\calibration and validation__MLR\Gopalkheda_365.xlsx")
df
```

```python
df1 = df[0:12418]
df2 = df[12418:]
df1
```

```python
df2
```

```python
model = M5Prime(use_pruning = True, use_smoothing = True)
model.fit(df1[["Q(t-1)"]].values,df1.Discharge.values)
predicted = model.predict(df2[["Q(t-1)"]].values)
print(r2_score(df2.Discharge, predicted), mean_squared_error(df2.Discharge, predicted,squared = False))
```
